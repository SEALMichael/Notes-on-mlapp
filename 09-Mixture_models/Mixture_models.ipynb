<h1><center> Mixture models </center></h1>

# 1. Basic
Model with hidden variables are known as **latent variable models (LVMs)**. Such models are harder to fit than models with no latent variables. However, they can have significant advantages, for two main reasons. 
* First, the LVMs often have fewer parameters than models that directly represent correlation in the visible space. 
* Second, the hidden variables in an LVM can serve as a **bottleneck**, which computes a compressed representation of the data.This forms the basis of unsurpevised learning.

The simplest form of LVM is when the latent variable is $z \in \lbrace 1,\ldots,K \rbrace$, representing a discrete latent state. We can use a discrete prior $p(z)=Cat(z|\vec{\pi})$,where $\vec{\pi}=(\pi_1,\ldots,\pi_K)$. The conditional distribution is written as $p(x|z=k)=p(x|\theta_k)$, the overall model is known as a mixture model. The joint distribution is as follows
\begin{align}
p(x,z=k|\theta) &=p(z=k)p(x|z=k) \\
                &=\pi_k p(x|\theta_k)
\end{align}
The likelihood function is
\begin{align}
p(D|\theta) &=\prod_{i=1}^N p(x) \\
            &=\prod_{i=1}^N \sum_{k=1}^K p(x,z=k|\theta) \\
            &=\prod_{i=1}^N \sum_{k=1}^K \pi_k p(x|\theta_k)
\end{align}

## 1.1 Some examples
## 1.1.1 Mixtures of Gaussian
The most widely used mixture model is the **mixture of Gaussian**, also called a **Gaussian mixture model (GMM)**. The model has the form
$$
p(x|\theta)=\sum_{k=1}^K\pi_k \mathcal{N}(x|\mu_k,\Sigma_k)
$$

### 1.1.2 Mixtures of multinoullis
Suppose our data consist of $D$-dimensional bit vectors. In this case ,an appropriate class-conditional density is a product of Bernoullis
\begin{align}
p(\vec{x}|z=k,\theta) &=\prod_{j=1}^d Ber(x_j|\theta_{jk}) \\
                      &=\prod_{j=1}^d \theta_{jk}^{x_j}(1-\theta_{jk})^{(1-x_j)}
\end{align}
The likelihood function is as follows
$$
p(D|\theta)=\prod_{i=1}^N \left[ \sum_{k=1}^K \prod_{j=1}^d Ber(x_j|\theta_{jk}) \right]
$$

## 1.2 Using mixture models for clustering
There are two main applications of mixture models. 
* The first is to use them as a **black-box** density model, $p(x)$. This can be useful for a variaty of tasks, such as outlier detection and creating generative classifiers.
* The second, and more common, application of mixture models is to use them for clustering.

We fist fit the mixture model and then compute $p(z_i|x_i,\theta)$, which represents the posterior probability that point i belongs to each cluster.
$$
p(z_i=k|x_i,\theta)= \frac{p(x_i|z_i=k,\theta)p(z_i=k|\theta)}{\sum_{k'=1}^K p(x_i|z_i=k',\theta)p(z_i=k'|\theta)}
$$
This procedure is called **soft clustering**, which output a distribution over $\lbrace 1,\ldots,K \rbrace$. We can also compute a MAP estimation, which called **hard clustering**
$$
\hat{z_i}=\underset{k}{argmax}\, p(z_i=k|x_i,\theta)
$$

# 2. Parameter estimation for mixture models
We have seen how to compute the posterior over the hidden variables given the observed variables, assuming the parameters are known. Now, we discuss how to learn the parameters.

In a directed graphical model, the posterior over the parameters factorizes if we have complete data and a factored prior, which make the parameter estimation very simple. Unfortunately, this is no longer true if we have hidden variables and/or missing data. The reason is as follows. If $z_i$ were observed, we see that $\theta_z \perp \theta_x |D$ and the posterior of $\theta_x,\theta_z$ will factorize. But since, in an LVM, the $z_i$ are hidden, the parameters are no longer independent, and the posterior does not factorize, which make it hard to compute.

## 2.1 Unidentifiability
The main problem with computing $p(\theta|D)$ for an LVM is that the posterior distribution may have multiple modes. Consider a GMM, if the $z_i$ were all observed, we would have a unimodal posterior for the parameters
$$
p(\theta|D)=Dir(\vec{\pi}|D)\prod_{k=1}^K NIW(\mu_k,\Sigma_k|D)
$$
which is a factorization form.

But now suppose the $z_i$ is hidden. In this case, the diffenrent value of $z_i$ is not identifiable, which means we can labels the clusters any way as we like. In this case, the posterior distribution may have multiple modes.

## 2.2 MAP estimate is non-convex
Consider the log-likelihood for an LVM
$$
log\,p(D|\theta)=\sum_{i=1}^N log\,\left[ \sum_{z_i} p(x_i,z_i|\theta) \right]
$$
Unfortunately, this objective is hard to maximize, since we cannot push the log inside the sum. The parameters are not factorized.

Now suppose the joint probability distribution $p(z,x|\theta)$ is in the exponential family ,which means
$$
p(x,z|\theta)=\frac{1}{Z(\theta)}exp \left[\theta^T \phi(x,z) \right]
$$

where $\phi(x,z)$ are sufficient statistics and $Z(\theta)$ is the normalization constant. Then the **complete data log likelihood** can be written as 
\begin{align}
log\,p(D|\theta) &=\sum_{i=1}^N log\, p(x_i,z_i|\theta) \\
                 &=\sum_{i=1}^N log\, \left[\frac{1}{Z(\theta)}exp \left[\theta^T \phi(x,z_i) \right] \right] \\
                 &=\theta^T\left(\sum_{i=1}^N \phi(x_i,z_i) \right)-NlogZ(\theta)
\end{align}
The complete data log likelihood is a concave function wrt $\theta$.

Now consider what happens when we have missing data. The **observed data log likelihood** is given by
\begin{align}
log\,p(D|\theta) &=\sum_{i=1}^N log\, \sum_{z_i} p(x_i,z_i|\theta) \\
                 &=\sum_{i=1}^N log\, \sum_{z_i} \left[\frac{1}{Z(\theta)}exp \left[\theta^T \phi(x,z_i) \right] \right] \\
                 &=\sum_{i=1}^N log\, \sum_{z_i} \left[exp \left(\theta^T \phi(x,z_i)\right) \right]-NlogZ(\theta)
\end{align}
The observed data log likelihood function is neither convex nor concave,which makes ML or MAP estimate difficult.

# 3. EM algorithm
EM algorithm exploits the fact that if the data were fully observed, the ML/MAP estimate would be easy to compute.In particular, EM is an iterative algorithm which alternates between inferring the missing values given the parameters(E step) and then optimizing the parameters given the "filled in" data(M step).

The **observed data log likelihood** is as follows
$$
l(\theta)=log\,p(D|\theta)=\sum_{i=1}^N log\,\sum_{z_i}p(x_i,z_i|\theta)
$$
we can define the **complete data log likelihood** to be
$$
l_c(\theta)=log\,p(D|\theta)=\sum_{i=1}^N log\,p(x_i,z_i|\theta)
$$
This connot be computed,since $z_i$ is unknown. We can define the **expected complete data log likelihood** 
$$
Q(\theta,\theta_{t-1})=E\left[l_c(\theta)\right]
$$
where t is the current iterative number. The expectation is taken wrt the estimated hidden variable $\hat{z}_i \sim p(\hat{z}_i|D,\theta_{t-1})$. In the E step, we compute $Q(\theta,\theta^{t-1})$ and in the M step ,we optimize the $Q$ function wrt $\theta$.
$$
\theta^t=\underset{\theta}{argmax}Q(\theta,\theta^{t-1})
$$
To perform MAP estimation, we modify the M step as follows
$$
\theta^t=\underset{\theta}{argmax}Q(\theta,\theta^{t-1})+log\,p(\theta)
$$

## 3.1 EM for GMMs
The expected complete data log likelihood is given below,The expectation is taken wrt the estimated hidden variable $\hat{z}_i \sim p(\hat{z}_i|D,\theta_{t-1})=Cat(\hat{z}_i|\vec{r_i})$.
\begin{align}
Q(\theta,\theta_{t-1}) &=E\left[l_c(\theta)\right] \\
                       &=E\left[\sum_{i=1}^N log\, p(x_i,\hat{z}_i=k|\theta)\right] \\
                       &=\sum_{i=1}^N E \left[ log\,\pi_k p(x_i|\theta_k) \right]  \\
                       &=\sum_{i=1}^N\sum_{k=1}^K r_{ik}\left[log\,\pi_k+log\,p(x_i|\theta_k)\right]
\end{align}
Please note that $\hat{z}_i$ were not chosen by **hard assign**, but rather by **soft assign**, which is a distribution $Cat(\hat{z}_i|\vec{r_i})$
The E step has the following simple form
$$
r_{ik}=p(z_i=k|D,\theta_{t-1})=\frac{p(z_i=k)p(x_i|\theta_k^{t-1})}{\sum_{k'=1}^K p(z_i=k')p(x_i|\theta_{k'}^{t-1})}
$$
In the M step, we optimize $Q$ wrt $\pi$ and $\theta_k$. For $\pi$, we obviously have
$$
\pi_k=\frac{\sum_{i=1}^N r_{ik}}{N}
$$
To derive the M step for the $\mu_k$ and $\Sigma_k$ terms, we look at the parts of $Q$ that depend on $\mu_k$ and $\Sigma_k$
\begin{align}
Q(\mu_k,\Sigma_k) &=\sum_{i=1}^N r_{ik}\left[log\,p(x_i|\theta_k)\right] \\
                  &=-\frac{1}{2}\sum_{i=1}^N r_{ik}\left[(x_i-\mu_k)^T\Sigma_{-1}(x_i-\mu_k)+d\,log\,(2\pi)+log|\Sigma_k| \right]
\end{align}
This is just a weighted version of the computing the MLEs of an MVN.One can show that the new parameter estimates are given by
\begin{align}
\mu_k  &=\frac{\sum_{i=1}^N r_{ik}x_i}{\sum_{i=1}^N r_{ik}} \\
\Sigma_k &=\frac{\sum_{i=1}^N r_{ik}(x_i-\mu_k)^T(x_i-\mu_k)}{\sum_{i=1}^N r_{ik}} \\
\end{align}

```python
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import normalize
from matplotlib.patches import Ellipse

group1=multivariate_normal([3,3],[[2.0,0.4],[0.4,0.5]]).rvs(size=500,random_state=42)
group2=multivariate_normal([0,0],[[0.3,0.2],[-0.2,0.6]]).rvs(size=300,random_state=42)

points=np.concatenate((group1,group2))
np.random.seed(42)
np.random.shuffle(points)

#Below is the EM algorithm implementation
groups=2    # The group num
points_num=points.shape[0]
r_i=np.zeros((points_num,groups))

pi=normalize(np.ones(groups).reshape(-1,1))
mu=[]
for i in range(groups):
    mu.append(points[i])
mu=np.array(mu)  # The center of each group
sigma=np.vstack(([np.identity(2)],)*groups)  #The covariance matrix of each group
iter=50
plot_freq=int(iter/9)+1

fig=plt.figure(figsize=(16,16))

def eigsorted(cov):
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    return vals[order], vecs[:,order]

for it in range(iter):
    if it%plot_freq==0 :
        ax=fig.add_subplot(3,3,int(it/plot_freq)+1,aspect='auto')
        ax.scatter(points[:,0],points[:,1],color='k')
        ax.scatter(mu[:,0],mu[:,1],color='r')
        ax.set_title("iter:%d"%it,fontsize=15)
        for g in range(groups):
            vals, vecs = eigsorted(sigma[g])
            theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
            w, h = 2 * 1.96 * np.sqrt(vals)
            ell = Ellipse(xy=mu[g],
                          width=w, 
                          height=h,
                          angle=theta, color='r',linewidth=3)
            ell.set_facecolor('None')
            ax.add_patch(ell)
            
    for g in range(groups):
        rv=multivariate_normal(mean=mu[g],cov=sigma[g])
        r_i[:,g]=rv.pdf(points)*pi[g]
    r_i=normalize(r_i,norm='l1')
    pi=normalize(np.mean(r_i,axis=0).reshape(-1,1))
    for g in range(groups):
        mu[g]=np.average(points,axis=0,weights=r_i[:,g])
        mat=np.array([np.matmul(x.reshape(2,-1),x.reshape(-1,2)) for x in (points-mu[g])])
        sigma[g]=np.average(mat,axis=0,weights=r_i[:,g])
```

## 3.2 MAP estimation
As usual, the MLE may overfit. The likelihood function is as follows
$$
p(D|\theta)=\prod_{i=1}^N \left[ \sum_{k=1}^K  p(x_i|z_i=k,\theta_k)p(z_i=k) \right]
$$
The overfit problem is particularly severe in the case of GMMs. To understand the problem, suppose for simplicity that $\mathbf{\Sigma}_k=\sigma_k^2\mathbf{I}$, and that $K=2$. It is possible to get an infinite likelihood by assigning one of the centers, say $\vec{\mu}_2$, to a single data, say $\vec{x}_1$, since then the $1$st term makes the following contribution to the likelihood
$$
\mathcal{N}(\vec{x}_1|\vec{\mu}_2,\sigma_2^2\mathbf{I})=\frac{1}{2\pi\sigma_2^2}e^0
$$
Hence we can drive this term to infinity by letting $\sigma_2 \rightarrow 0$. **So the observed data log likelihood might increase to infinite.** The EM monotonically increases the observed data log likelihood until
it reaches a local maximum (or saddle point, although such points are usually unstable).

An easy solution to this is to perform MAP estimation. The new objective function is the expected complete data log-likelihood plus the log prior
$$
Q(\theta,\theta_{t-1})=E \left[ l_c(\theta) \right ] +log\,p(\theta)
$$
The expectation is taken wrt the estimated hidden variable $\hat{z}_i \sim p(\hat{z}_i|D,\theta_{t-1})$

## 3.3 K-means algorithm
There is a popular variant of the EM algorithm for GMMs known as the **k-means algorithm**. Consider a GMM in which we make the following assumptions $\Sigma_k=\sigma^2\mathbf{I}_D$ is fixed and $\pi_k=\frac{1}{K}$, so only the cluster centers, $\mu_k$ have to be estimated. We also use a MAP estimation during the E step
$$
\hat{z}_i=\underset{k}{argmax}\,p(z_i=k|x_i,\theta^{t-1})
$$
This is called **hard EM**, since we are making a hard assignment of points to clusters instead of compute the expectation wrt $\hat{z}_i$. Since we assumed an equal spherical covariance matrix for each cluster, the most probable cluster for $x_i$ can be computed by finding the nearest prototype:
$$
\hat{z}_i=\underset{k}{argmax} ||x_i-\mu_k||_2^2
$$
The K-means algorithm can be interpreted as a solution to the data compression. Suppose we want to perform lossy compression of some real-valued vectors,$\vec{x_i} \in R^D$. The basic idea is to replace each vector with a discrete symbol $z_i \in \lbrace 1,\ldots,K \rbrace $ which is an index into a **dictionary** of $\mathbf{K}$ prototypes, $\mu_k \in R^D$.Each data vector is encoded by using the index of the most similar prototype, where similarity is measured in terms of Euclidean distance
$$
encode(\vec{x}_i)=\underset{k}{argmin}||\vec{x}_i-\vec{\mu}_k||^2
$$
We can define a cost function that measures the quality of a codebook by computing the **reconstruction error** as follows
$$
J(\vec{\mu}|K,\mathbf{X}) =\frac{1}{N}\sum_{i=1}^N \|\vec{x}_i-decode(encode(\vec{x}_i))\|^2
$$
where $decode(k)=\vec{\mu}_k$. The K-means algorithm can be thought of as a simple iterative scheme for minimizing this objective.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

n_samples=1500
X,y=make_blobs(n_samples=n_samples,random_state=42)

y_pred=KMeans(n_clusters=2,random_state=42,init='k-means++').fit_predict(X)

plt.figure(figsize=(16,16))
plt.subplot(2,2,1)
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.title("Incorrect Number of Blobs",fontsize=15)

transformation = [[0.60834549, -0.83667341], [-0.70887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred=KMeans(n_clusters=3,random_state=42,init='k-means++').fit_predict(X_aniso)

plt.subplot(2,2,2)
plt.scatter(X_aniso[:,0],X[:,1],c=y_pred)
plt.title("Anisotropicly Distributed Blobs",fontsize=15)

X_varied,y_varied=make_blobs(n_samples=n_samples,
                            cluster_std=[1.0,2.5,0.5],
                            random_state=42)
y_pred=KMeans(n_clusters=3,random_state=42,init='k-means++').fit_predict(X_varied)

plt.subplot(2,2,3)
plt.scatter(X_varied[:,0],X_varied[:,1],c=y_pred)
plt.title("Unequal Variance",fontsize=15)

X_unbalance=np.vstack((X[y==0][:500],X[y==1][:100],X[y==2][:10]))
y_pred=KMeans(n_clusters=3,random_state=42,init='k-means++').fit_predict(X_unbalance)

plt.show()
```

# 4. Theoretical basis for EM
Consider an arbitrary distribution $q_i(z_i)$ over the hidden variables. The expected complete data log likelihood wrt to $z_i \sim q_i(z_i)$ is
$$
Q(\vec{\theta},q_{i=1}^N)=\sum_{i=1}^N \left[ \sum_{z_i}q_i(z_i)\,log\,p(x_i,z_i|\vec{\theta}) \right]
$$
The observed data log likelihood can be written as follows
\begin{align}
l(\vec{\theta})  &=\sum_{i=1}^N\left[log\,p(\vec{x}_i|\vec{\theta})\right]  \\
                 &=\sum_{i=1}^N\left[\sum_{z_i} \left[q_i(z_i)\,log\,p(\vec{x}_i|\vec{\theta})\right]\right] \\
                 &=\sum_{i=1}^N\left[\sum_{z_i} \left[q_i(z_i)\,log\,\frac{p(\vec{x}_i,z_i|\vec{\theta})}{p(z_i|\vec{x}_i,\vec{\theta})}\right]\right] \\
                 &=\sum_{i=1}^N\left[\sum_{z_i} \left[q_i(z_i)\,log\,\frac{p(\vec{x}_i,z_i|\vec{\theta})}{q_i(z_i)}-q_i(z_i)\,log\,\frac{p(z_i|\vec{x}_i,\vec{\theta})}{q_i(z_i)}\right]\right] \\
                 &=\mathcal{L}(q_{i=1}^N,\vec{\theta})+\sum_{i=1}^N \text{KL}(q_i \| p(z_i|\vec{x}_i,\vec{\theta}))
\end{align}
where we have defined
\begin{align}
\mathcal{L}(q_{i=1}^N,\vec{\theta}) &=\sum_{i=1}^N\left[\sum_{z_i} \left[q_i(z_i)\,log\,\frac{p(\vec{x}_i,z_i|\vec{\theta})}{q_i(z_i)} \right]\right] \\
                                    &=Q(\vec{\theta},q_{i=1}^N)+\sum_{i=1}^N \mathbb{H}[q_i]\\
\end{align}
Recall that the **Kullback-Leibler divergence** satisfies $\text{KL}(q_i \| p(z_i|\vec{x}_i,\vec{\theta})) \geq 0$, with equality iff $q_i=p(z_i|\vec{x}_i,\vec{\theta})$. 

* In the E step, the low bound $\mathcal{L}(q_{i=1}^N,\vec{\theta})$ is maximized wrt $q_{i=1}^N$ while holding $\vec{\theta}^{t-1}$ fixed.The solution to this maximization problem is easily seen by noting that the value of $l(\vec{\theta})$ does not depend on $q_{i=1}^N$ and so the largest value of $\mathcal{L}(q_{i=1}^N,\vec{\theta})$  will occur when the **Kullback-Leibler divergence** vanishes, in other words when $q_i=p(z_i|\vec{x}_i,\vec{\theta}^{t-1})$. In this case, the lower bound will equal the log likelihood.
* In the M step, the distribution $q_{i=1}^N$ is held fixed and the lower bound $\mathcal{L}(q_{i=1}^N,\vec{\theta})$ is maximized with respect to $\vec{\theta}$ to give $\vec{\theta}^t$,which will increase the lower bound $\mathcal{L}$.

We can find that EM monotonically increases the observed data log likelihood until it reaches a local optimum. We have
\begin{align}
l(\vec{\theta}^t) &\geq \mathcal{L}\left(q_{i=1}^N=p(z_i|\vec{x}_i,\vec{\theta}^{t-1}),\vec{\theta}^t\right) \\
                  &\geq \mathcal{L}\left(q_{i=1}^N=p(z_i|\vec{x}_i,\vec{\theta}^{t-1}),\vec{\theta}^{t-1}\right)  \\
                  &=l(\vec{\theta}^{t-1})
\end{align}
